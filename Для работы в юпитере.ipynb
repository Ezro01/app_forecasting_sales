```python
import pandas as pd 

df = pd.read_csv('Dataframe_500_tovars_magazins.csv', parse_dates=['Дата'])
df_1 = pd.read_csv('train_df.csv', parse_dates=['Дата'])
```

```python
print(df['Дата'].min())
print(df['Дата'].max())
```

```python
print(df_1['Дата'].min())
print(df_1['Дата'].max())
```

```python
date_stop = df_1['Дата'].max() + pd.Timedelta(days=7)
print(date_stop)
```

```python
df_res = df[df['Дата'] > df_1['Дата'].max()]
df_res = df_res[df_res['Дата'] < date_stop]
print(df_res['Дата'].min())
print(df_res['Дата'].max())
df_res.to_csv('test_df.csv', index=False)
```

# ПРоверка совпадений 

```python
import pandas as pd 

df_train = pd.read_csv('train_df.csv', parse_dates=['Дата'])
df_test = pd.read_csv('test_df.csv', parse_dates=['Дата'])

df_train_1 = set(zip(df_train['Магазин'], df_train['Товар']))
df_test_1 = set(zip(df_test['Магазин'], df_test['Товар']))

# Находим пересечение
common_pairs = df_train_1 & df_test_1

print(f"Всего пар в df_train_1: {len(df_train_1)}")
print(f"Всего пар в df_test_1: {len(df_test_1)}")
print(f"Совпадающих пар: {len(common_pairs)}")
```
